```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../train.csv')

# Create a count plot
plt.figure(figsize=(8, 5))  # Set the figure size
sns.countplot(x='label_boolean', data=df, palette='viridis')

# Add labels and title
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Class Distribution in Dataset')
plt.xticks([0, 1], ['False', 'True'])  # Assuming 0 = False, 1 = True

# Display the plot
plt.show()
```

```python
df[['metaphorID','label_boolean']].value_counts()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_counts = df[['metaphorID','label_boolean']].value_counts().reset_index(name='count')

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='metaphorID', y='count', hue='label_boolean', data=df_counts, palette='viridis')
plt.title('Counts of True and False Labels by MetaphorID')
plt.xlabel('Metaphor ID')
plt.ylabel('Count')
plt.show()
```
